```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_multiple(x_label, y_label, title, interval, lr, als):
    plt.figure(figsize=(4, 6), dpi=100)
    data = pd.read_csv(f'Ouput/cost_lr_{lr}.csv')
    for a in als:
        plt.plot(range(interval[0], interval[1]), data[a][interval[0]:interval[1]], label=a)
    plt.xlim(interval[0]-1, interval[1])
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(f'{title} (learning rate {lr} )')
    plt.legend()
    plt.show()
```

```python
als = ['gd','bgd','acc','bac','nt','bnt']
plot_multiple('count', 'loss function', 'loss function vs count',interval=[0, 10], lr=0.1, als=als)
```
