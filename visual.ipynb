```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_multiple(x_label, y_label, title, interval, lr, als):
    plt.figure(figsize=(6, 4), dpi=100)
    for a in als:
        data = pd.read_csv(f'output_cost_lr_{lr}_{a}.csv')
        plt.plot(range(interval[0], interval[1]), data['0'][interval[0]:interval[1]], label=a)
    plt.xlim(interval[0]-1, interval[1])
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(f'{title} (learning rate {lr} )')
    plt.legend()
    plt.show()
```

```python
als = ['bgd','bac','bnt']
plot_multiple('count', 'loss function', 'loss function vs count',interval=[0,200],lr=0.5,als=als)
```
